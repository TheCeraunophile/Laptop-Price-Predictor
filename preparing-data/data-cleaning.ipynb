```python
df.loc[df['amount'] =='nmwjwd', 'amount'] = np.NaN
df.loc[df['amount'] =='?', 'amount'] = np.NaN

df = df.dropna(subset=['amount'])
df.T
```

```python
import re
def get_value(x):
    return re.findall('^[0-9]*.[0-9]*', x)

result = map(get_value, df['amount'])

result = list(result)
result = np.array(result)

result = result.reshape(result.shape[0] * result.shape[1])

df['amount'] = result

df['amount'] = df['amount'].astype(float)
```
