```python
import selenium
from selenium import webdriver
import pandas as pd 
import numpy as np
```

```python
URL = 'https://torob.com/browse/99/%D9%84%D9%BE-%D8%AA%D8%A7%D9%BE-%D9%88-%D9%86%D9%88%D8%AA-%D8%A8%D9%88%DA%A9-laptop/?stock_status=new&available=true'

driver = webdriver.Chrome()

driver.get(URL);
```

```python
records = driver.find_elements('xpath', '//a[@class="jsx-fa8eb4b3b47a1d18"]')
len(records)
```

```python
record_links = []
for laptop in records:
    record_links.append(laptop.get_attribute('href'))
```

```python
arr = np.asarray(record_links)
pd.DataFrame(arr).to_csv('laptop_links.csv') 
```

```python
df = pd.read_csv('laptop_links.csv')

df['0']
```

```python
record_links = df['0'].to_numpy()
```

```python
def get_data(start, end):
    i = start
    j = 0
    for link in record_links[start:end]:
        print('iteration   ', i)
        driver.get(link)
        features = driver.find_elements('xpath', '//div[@class="jsx-5b5c456cc255c2dc"]')
        value = []
        feature = []

        amount = driver.find_element('xpath', '//div[@class="jsx-63b317fab2efbae ellipsis"]')
        feature.append('amount')
        value.append(amount.text.split('\n')[1].split(' ')[0])

        for element in features:
            step1 = element.text.split('\n')
            if len(step1) == 1:
                continue
    #         _feature, _value = step1
    #         value.append(_value)
    #         feature.append(_feature)
            value.append(step1[1])
            feature.append(step1[0])

        df.at[i,feature] = value
        ndf.at[j,feature] = value
        i = i + 1
        j = j + 1
    df.to_csv('from ' + str(0) +  '     to ' + str(end) +  '.csv')
    ndf.to_csv('from ' + str(start) +  '     to ' + str(end) +  '.csv')
```

```python
df = pd.DataFrame()
get_data(0, 200)
print(df.shape)
df
```

```python
ndf = pd.DataFrame()
get_data(200, 400)
print(df.shape)
df
```

```python
import requests
from bs4 import BeautifulSoup

def new_get_data(start, end):
    i = start
    j = 0
    for link in record_links[start:end]:
        print('iteration   ', i)
        result = requests.get(link)
        soup = BeautifulSoup(result.content, 'html.parser')


        amount = soup.find_all('div', class_="jsx-63b317fab2efbae buy_box_text ellipsis")
        if len(amount) == 0:
            i = i + 1
            j = j + 1
            continue
        
        features = soup.find_all('div', class_="jsx-5b5c456cc255c2dc detail-title")
        values = soup.find_all('div', class_="jsx-5b5c456cc255c2dc detail-value")

        result = requests.get(record_links[i])


        
        amount = amount[1].get_text().split(' ')[0]

        feature = []
        value = []

        feature.append('amount')
        value.append(amount)

        for k in range(len(features)):
            value.append(values[k].get_text())
            feature.append(features[k].get_text())

        df.at[i,feature] = value
        ndf.at[j,feature] = value
        i = i + 1
        j = j + 1
    df.to_csv('from ' + str(0) +  '     to ' + str(end) +  '.csv')
    ndf.to_csv('from ' + str(start) +  '     to ' + str(end) +  '.csv')
```
